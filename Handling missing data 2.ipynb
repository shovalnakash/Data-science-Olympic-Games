**we have a problem in column 'Place of birth', the uniqe values are 3624 values.
so we will fix it by split the country and city, using only the country the athlete came from:**

```python
pob= df['Place of birth']
pob
```

```python
for i in range(0,len(pob)):
    newVal= re.split("[()]", pob[i])
    try:
        pob[i]= newVal[1]
    except:
        pob[i]= np.nan
df['Place of birth'] = pob
```

```python
df.describe(include='all')
```

**#now we can see the unique values of column 'Place of birth' is 19, a number we can work with.**

**#we need to drop the columns that dont give us important information for our research**

```python
df.drop(['Name', 'Country', 'Date of birth', 'Event'], axis=1, inplace=True)
df.dropna(axis=0, inplace= True)
```

```python
df.reset_index(drop=True, inplace=True) #we will reset again after delete
df.head()
```

# data treatment- letters to numeric

**# check which category need our treatment and see what kind o possible values we have**

```python
print('Age:', sorted(list(df['Age'].unique())))
print('Height, cm:', sorted(list(df['Height, cm'].unique())))
print('Weight:', sorted(list(df['Weight'].unique())))
print('Sex:', sorted(list(df['Sex'].unique())))
print('Place of birth:', sorted(list(df['Place of birth'].unique())))
print('Gold:', sorted(list(df['Gold'].unique())))
print('Silver:', sorted(list(df['Silver'].unique())))
print('Bronze:', sorted(list(df['Bronze'].unique())))
print('Total:', sorted(list(df['Total'].unique())))
print('Sport:', sorted(list(df['Sport'].unique())))
```

we can see, that we have 3 categories that need our treatment:
### 1. change the values of sex to binary (men/female)
### 2. change the sport to numeric values (1-34)
### 3. change the 'place of birth column to numeric values (1-194) 

```python
df= pd.get_dummies(df, columns=['Sex'], prefix=['Sex'])
```

```python
numericSport = dict(zip(range(1,37),df.Sport.unique()))
numericSport = dict(zip(numericSport.values(), numericSport.keys()))
df['Sport'] = df['Sport'].map(numericSport).fillna(df['Sport'])
```

```python
country_labels = df["Place of birth"].astype('category').cat.categories.to_list()
rep_map={x:y for x,y in zip(country_labels,list(range(0,len(country_labels)+1)))}
df["Place of birth"].replace(rep_map, inplace=True)
```
