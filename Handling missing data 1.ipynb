## Handling missing data and duplicates

###### 1# cleaning the Dataframe from null objects

###### 2# cleaning duplicates objects

```python
copydf= OlympicDf
```

```python
copydf.dropna(axis=0, thresh=14, inplace= True)
```

```python
copydf.isnull().sum()
```

```python
copydf.isnull().sum().sum()
```

###### no missing data anymore

```python
copydf.head()
```

```python
copydf.drop_duplicates(subset=['Name'], inplace= True)
```

```python
len(copydf)
```

###### no duplicates rows

after handling the data we will save all changes in new csv:

```python
copydf.to_csv("All London 2012 athletes cleaning.csv", index=False)
```
